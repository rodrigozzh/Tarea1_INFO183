# 1) Transformada de Fourier de coseno y seno

Considere un vector de tiempo con duración 10 segundos y frecuencia de muestreo 100 Hz

Calcule la transformada de Fourier de $x$

```python
t = np.arange(0.0, 10.0, step=0.01)
x = np.cos(2.0*np.pi*t)
plt.figure()
plt.plot(t, x);
```

Muestre graficamente
- La parte real de la transformada
- La parte imaginaria de la transformada
- El valor absoluto de la transformada
- El ángulo de la transformada


Indique el valor absoluto y el ángulo para la frecuencia fundamental de la señal (1Hz)

```python
fs = 100 
st = 1.0 / fs 
t = np.arange(0.0, 10.0, step=0.01)
signal1 = np.cos(2.0*np.pi*t) #SEÑAL COS
signal2 = np.sin(2.0*np.pi*t) #SEÑAL SEN
Array_Abs= np.abs(fftpack.fft(signal1))
Array_angl = np.angle(fftpack.fft(signal1))
_, axes = plt.subplots(5, 2)

#SEÑALES COS Y SIN
axes[0][0].set_title("Signal 1")
axes[0][0].grid()
axes[0][0].plot(t, signal1, 'b-')

axes[0][1].set_title("Signal 2")
axes[0][1].grid()
axes[0][1].plot(t, signal2, 'b-')

#PARTES REALES DE LAS SEÑALES COS Y SIN
axes[1][0].set_title("Signal(Cos) Real")
axes[1][0].plot(fftpack.fftfreq(d=0.01,n=len(t)), np.real(fftpack.fft(signal1)), linewidth=2); 
axes[1][0].set_xlim([-3, 3])

axes[1][1].set_title("Signal(Sin) Real")
axes[1][1].plot(fftpack.fftfreq(d=0.01,n=len(t)), np.real(fftpack.fft(signal2)), linewidth=2);

#PARTES IMAGINARIAS DE LAS SEÑALES COS Y SIN
axes[2][0].set_title("Signal(Cos) Imaginaria")
axes[2][0].plot(fftpack.fftfreq(d=0.01,n=len(t)), np.imag(fftpack.fft(signal1)), linewidth=2);

axes[2][1].set_title("Signal(Sin) Imaginaria")
axes[2][1].plot(fftpack.fftfreq(d=0.01,n=len(t)), np.imag(fftpack.fft(signal2)), linewidth=2);
axes[2][1].set_xlim([-3, 3])

#PARTES VALORES ABSOLUTOS DE LAS SEÑALES COS Y SIN
axes[3][0].set_title("Signal(Cos) Abs")
axes[3][0].plot(fftpack.fftfreq(d=0.01,n=len(t)), np.abs(fftpack.fft(signal1)), linewidth=2);
axes[3][0].set_xlim([-3, 3])

axes[3][1].set_title("Signal(Sin) Abs")
axes[3][1].plot(fftpack.fftfreq(d=0.01,n=len(t)), np.abs(fftpack.fft(signal2)), linewidth=2);
axes[3][1].set_xlim([-3, 3])

#PARTES ANGULOS DE LAS SEÑALES COS Y SIN
axes[4][0].set_title("Signal(Cos) Angle")
axes[4][0].plot(fftpack.fftfreq(d=0.01,n=len(t)), np.angle(fftpack.fft(signal1)), linewidth=2);

axes[4][1].set_title("Signal(Sin) Angle")
axes[4][1].plot(fftpack.fftfreq(d=0.01,n=len(t)), np.angle(fftpack.fft(signal2)), linewidth=2);


plt.subplots_adjust(hspace= 0.5)
plt.subplots_adjust(bottom=0.3, right=2, top=5)

# OBTENER EL VALOR ABSOLUTO Y ANGULO FUNDAMENTAL PARA 1[Hz]
Array = fftpack.fftfreq(d=0.01,n=len(t))
largo=len(Array)
for i in range(0, largo):
    if Array[i] == 1.0:
        indice = i;
print("Para valor absoluto y el ángulo para la frecuencia fundamental de la señal (1Hz) ")
print(">>> Valor Absoluto fundamental: ", Array_Abs[indice])
print(">>> Ángulo fundamental: ", Array_angl[indice])

```

# 2) Análisis de señal estacionaria

Considere la señal estacionaria con 500 muestras y frecuencia de muestreo 50 Hz guardada en `signal.dat`

Responda
1. ¿Cuál es la duración temporal de la señal?

2. ¿Cuál es la mayor frecuencia que podemos estudiar usando Fourier en este escenario?



Calcule la transformada de Fourier de este señal y grafique
- La parte real de la transformada
- La parte imaginaria de la transformada
- El valor absoluto de la transformada
- El ángulo de la transformada
    
En base a estas visualizaciones y lo observado en el punto anterior indique la fórmula que generó esta señal

```python
plt.close('all'); fig, axes = plt.subplots(6, figsize=(6, 9))

fs = 50 
st = 1.0 / fs 
t, dt = np.linspace(0, 10, num=500, retstep=True);
signal = np.loadtxt('signal.dat')
f0=2


axes[0].set_title("Señal data")
axes[0].plot(signal);

axes[1].set_title("Signal real")
axes[1].plot(np.sort(fftpack.fftfreq(n=len(t), d=dt)), np.real(fftpack.fftshift(fftpack.fft(signal))))
axes[1].set_xlim([-10, 10])

axes[2].set_title("Signal imag")
axes[2].plot(np.sort(fftpack.fftfreq(n=len(t), d=dt)), np.imag(fftpack.fftshift(fftpack.fft(signal))))
axes[2].set_xlim([-10, 10])

axes[3].set_title("Signal abs")
axes[3].plot(np.sort(fftpack.fftfreq(n=len(t), d=dt)), np.absolute(fftpack.fftshift(fftpack.fft(signal))))
axes[3].set_xlim([-10, 10])

axes[4].set_title("Signal angle")
axes[4].plot(np.sort(fftpack.fftfreq(n=len(t), d=dt)), np.angle(fftpack.fftshift(fftpack.fft(signal))))
axes[4].set_xlim([-10, 10])

axes[5].set_title("FUNCION ESTIMADA")
axes[5].plot(np.cos(2.5*np.pi*f0*t) - 0.6*np.cos(2.5*np.pi*2*f0*t +np.pi/4))

plt.subplots_adjust(hspace= 0.3)
plt.subplots_adjust(bottom=0.7, right=4, top=5)
```

# 3) Estimación de frecuencia fundamental en señal de audio

Considere la siguiente señal de audio no estacionaria

```python
import soundfile as sf
from IPython.display import Audio

fig, ax = plt.subplots(figsize=(8, 6))
data, sample_rate = sf.read("data/DPSAU.ogg")
time = np.linspace(0.0, len(data)/sample_rate, num=len(data))
ax.plot(time, data);
Audio(data, rate=sample_rate)
```

- Escriba una rutina en Python usando `scipy.fftpack.fft` que estime la frecuencia fundamental de esta señal **en función del tiempo** 
- En base a lo anterior indique los 8 tonos musicales utilizados en esta melodia 

Ref: https://es.wikipedia.org/wiki/Frecuencias_de_afinaci%C3%B3n_del_piano

```python
import scipy.fftpack as sfft
from style import *

fig, ax = plt.subplots(8,1,figsize=(15,25))
limites =[0.5,2.8,5.5,7.6,10.2,12.4,14.9,17.3] #limites estimados del grafico anterior, para crear los intervalos.
for i in range(8):
    signal=sfft.fft(data [int(limites[i]*sample_rate): int(limites[i]*sample_rate)+(int(1.5*sample_rate))])
    freq=sfft.fftfreq(int(1.5*sample_rate),1./sample_rate)
    ax[i].plot(freq,abs(signal))
    ax[i].set_xlim([0.,2000.])
    title="Tiempo entre "+str(limites[i])+" y "+str(limites[i]+1.5)+" segundos"
    ax[i].set_title(title)
```
